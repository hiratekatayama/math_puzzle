```python
def ppr(a,b):
    import numpy as np
    cnt = 0
    for i in range(len(a)):
        if len(b) == 4:
            if a[i] == b[3-i]:
                cnt += 1
                print(cnt,a,b)
        if cnt == 4:
            print(a[i],b[i])
```

```python
import numpy as np

for i in range(1000,10000):
    a = str(i)
    for j in range(1,4):
        b = str(int(a[j:])*int(a[:j]))
        ppr(a,b)
```
